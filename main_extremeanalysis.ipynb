This code  generete extreme analysis images to define the best distribution:
    
First cel = Libraries needed
Second cel = Operational Functions and Main Function

```python
import pandas as pd 
from pyextremes import EVA,get_extremes
from pyextremes.plotting import plot_extremes
import matplotlib.pyplot as plt
import mikeio
import os
import warnings

warnings.filterwarnings("ignore")
```

```python
######################################## Operational Functions #############################################

#Create series for analysis
def get_dfseries(df,variable):
    if variable in ['Wind Speed']:
        series=df.loc[:][variable]
    elif variable in ['Current Speed']:
        series=df.loc[:][variable]
    elif variable in ['Hm0']:
        series=df.loc[:][variable]
    elif variable in ['Hmax']:
        series=df.loc[:][variable]
    elif variable in ['Air Temperature']:
        series=df.loc[:][variable]
    elif variable in ['Sea Water Temperature']:
        series=df.loc[:][variable]
    elif variable in ['Water Level']:
        series=df.loc[:][variable]
    return series

#Create results folders
def create_folder_result_ominidirectional(point,variable,depth):

    if variable in ['Wind Speed','Current Speed']:
        variable = variable.split(' ')[0]

    if str(depth) =='nan':
        path_results='C:/Users/test/Desktop/Scripts/Extreme/Results/'+point+'/'+variable+'/Omnidirectional/'
        if not os.path.exists(path_results):    
            os.makedirs(path_results)
    else:
        path_results='C:/Users/test/Desktop/Scripts/Extreme/Results/'+point+'/'+variable+'/'+depth+'/Omnidirectional/'
        if not os.path.exists(path_results):    
            os.makedirs(path_results)     

    return path_results

def create_folder_result_directional(point,variable,direction,depth):

    if variable in ['Wind Speed','Current Speed']:
        variable = variable.split(' ')[0]

    if str(depth) =='nan':
        path_results='C:/Users/test/Desktop/Scripts/Extreme/Results/'+point+'/'+variable+'/Directional/dfs_'+direction+'/'
        if not os.path.exists(path_results):    
            os.makedirs(path_results)
    else:
        path_results='C:/Users/test/Desktop/Scripts/Extreme/Results/'+point+'/'+variable+'/'+depth+'/Directional/dfs_'+direction+'/'
        if not os.path.exists(path_results):    
            os.makedirs(path_results)
            
    return path_results

#Save Extreme plot
def save_extremes(series,threshold_selected,time_selected,path_results):
    Extremes=get_extremes(series,"POT",threshold=threshold_selected, r=time_selected)
    Extremes_plot=plot_extremes(ts=series,extremes=Extremes,extremes_method='POT',
                                extremes_type='high',threshold=threshold_selected,r=time_selected)
    Extremes_Fig=Extremes_plot[0].figure
    Extremes_Fig.savefig(path_results+'ExtremesPoints',bbox_inches='tight')

# Adjust different models distribution
def model_analysis(series,threshold_selected,time_Selected,path_results):
############################ Adjust model EXPON ##################
    modelExpon=EVA(series)
    modelExpon.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    #print(modelExpon)
    modelExpon.fit_model(distribution='expon')
    summaryExpon=modelExpon.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modelExpon.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelExpon=modelExpon.plot_diagnostic(alpha=0.95)[0]
    fig_modelExpon.savefig(path_results+'Expon')
    summaryExpon.to_csv(path_results+'Expon.txt',sep='\t',index='False')

############################ Adjust model GENPARETO ##################
    modelGenpareto=EVA(series)
    modelGenpareto.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    modelGenpareto.fit_model(distribution='genpareto')
    summaryGenpareto=modelGenpareto.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modelGenpareto.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelGenpareto=modelGenpareto.plot_diagnostic(alpha=0.95)[0]
    fig_modelGenpareto.savefig(path_results+'Genpareto')
    summaryGenpareto.to_csv(path_results+'Genpareto.txt',sep='\t',index='False')

############################ Adjust model TRUNCPARETO ##################
    modeltruncpareto=EVA(series)
    modeltruncpareto.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    modeltruncpareto.fit_model(distribution='truncpareto')
    summaryTruncpareto=modeltruncpareto.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modeltruncpareto.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelTruncpareto=modeltruncpareto.plot_diagnostic(alpha=0.95)[0]
    fig_modelTruncpareto.savefig(path_results+'Truncpareto')
    summaryTruncpareto.to_csv(path_results+'Truncpareto.txt',sep='\t',index='False')

############################ Adjust model PEARSON3 ##################

    modelPearson3=EVA(series)
    modelPearson3.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    modelPearson3.fit_model(distribution='pearson3')
    summaryPearson3=modelPearson3.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modelPearson3.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelPearson3=modelPearson3.plot_diagnostic(alpha=0.95)[0]
    fig_modelPearson3.savefig(path_results+'Pearson3')
    summaryPearson3.to_csv(path_results+'Pearson3.txt',sep='\t',index='False')

############################ Adjust model GENEXTREME ##################
    
    modelGenextreme=EVA(series)
    modelGenextreme.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    modelGenextreme.fit_model(distribution='genextreme')
    summaryGenextreme=modelGenextreme.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modelGenextreme.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelGenextreme=modelGenextreme.plot_diagnostic(alpha=0.95)[0]
    fig_modelGenextreme.savefig(path_results+'Genextreme')
    summaryGenextreme.to_csv(path_results+'Genextreme.txt',sep='\t',index='False')
    
############################ Adjust model PARETO ##################
    
    modelPareto=EVA(series)
    modelPareto.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    modelPareto.fit_model(distribution='pareto')
    summaryPareto=modelPareto.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modelPareto.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelPareto=modelPareto.plot_diagnostic(alpha=0.95)[0]
    fig_modelPareto.savefig(path_results+'Pareto')
    summaryPareto.to_csv(path_results+'Pareto.txt',sep='\t',index='False')
        
############################ Adjust model LOGNORM ##################
    
    modelLognorm=EVA(series)
    modelLognorm.get_extremes("POT",threshold=threshold_selected, r=time_Selected)
    modelLognorm.fit_model(distribution='lognorm')
    summaryLognorm=modelLognorm.get_summary(
        return_period=[1,5, 10, 20, 50, 100],
        alpha=0.95)
    modelLognorm.plot_diagnostic(alpha=0.95)

    #Save Results
    fig_modelLognorm=modelLognorm.plot_diagnostic(alpha=0.95)[0]
    fig_modelLognorm.savefig(path_results+'Lognorm')
    summaryLognorm.to_csv(path_results+'Lognorm.txt',sep='\t',index='False')

######################################## Main Function ################################################################
# Read excel
df_info=pd.read_csv('C:/Users/test/Desktop/Scripts/Extreme/Extreme_Analysis.csv',sep=';')
time_selected="36h" 

for d in range(len(df_info)):

    if df_info.loc[d]['Index'] == 1:
            
            if df_info.loc[d]['Direction'] =='Omnidirectional':
                path_input='C:/Users/test/Desktop/Scripts/Extreme/Omnidirectional/'
                path_result=create_folder_result_ominidirectional(df_info.loc[d]['Points'],df_info.loc[d]['Variable'],df_info.loc[d]['Depth'])
                ds=mikeio.read(path_input+df_info.loc[d]['File']+'.dfs0')
                df=ds.to_dataframe()
                print(ds)
                series=get_dfseries(df,df_info.loc[d]['Variable'])
                save_extremes(series,df_info.loc[d]['Threshold'],time_selected,path_result)
                model_analysis(series,df_info.loc[d]['Threshold'],time_selected,path_result)
            else:
                path_input='C:/Users/test/Desktop/Scripts/Extreme/Directional/'+df_info.loc[d]['File']+'/'
                path_result=create_folder_result_directional(df_info.loc[d]['Points'],df_info.loc[d]['Variable'],df_info.loc[d]['Direction'],df_info.loc[d]['Depth'])
                direction='dfs_'+df_info.loc[d]['Direction']
                ds=mikeio.read(path_input+direction+'.dfs0')
                df=ds.to_dataframe()
                series=get_dfseries(df,df_info.loc[d]['Variable'])
                save_extremes(series,df_info.loc[d]['Threshold'],time_selected,path_result)
                model_analysis(series,df_info.loc[d]['Threshold'],time_selected,path_result)

    else:
        pass
```
